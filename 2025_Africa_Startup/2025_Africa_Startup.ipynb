- 1. Importing Libraries, Loading Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np

df = pd.read_csv(r'Dataset\2025_Africa_Startup&VC_Landscape_Preview_for_Founders_Investors_LPs_DFIs_and_Public_Officials_Guests.csv')
df.head()
```

- 2. Data Cleanup and Feature Engineering

```python
# Step 1: Clean 'phone_number' field
df['phone_number'] = df['phone_number'].fillna(0).astype('int64').astype(str)
df['phone_number'] = df['phone_number']

# Replace all null values with "No Input"
df.fillna("No Input", inplace=True)

# Step 2: Rename 'created_at' to 'date_created' and Convert to datetime
df = df.rename(columns={
    'created_at' : 'date_created',
    'What best describes you?' : 'describe_attendee', 
    'What best describes your organization?' : 'describe_organization',
    'If Founder: Are You Currently Raising Funds?' : 'founder_raising_fund?',
    'Startups: What region does your company focus on?' : 'startup_company_region',
    'What are your expectations for this conference?' : 'conference_expectation',
    'Company Name' : 'company_name',
    'Company Email' : 'company_email',
    'Company Website' : 'company_website',
    'Company Contact' : 'company_contact',
    'Short Bio' : 'short_bio',
    })
df['date_created'] = pd.to_datetime(df['date_created'], errors='coerce')
df['date_created'] = pd.to_datetime(df['date_created'].dt.strftime('%Y-%m-%d %H:%M:%S'))


# Step 4: Drop Irrelevant Columns (columns with no data at all)
irrelevant_cols = ['coupon_code', 'amount', 'amount_tax', 'amount_discount', 'custom_source', 'currency' 'eth_address', 'solana_address', 
                   'survey_response_rating', 'survey_response_feedback']
df = df.drop(columns=irrelevant_cols, errors='ignore')

df.to_excel('Cleaned_Dataset/Cleaned_Africa_Startup_2025.xlsx')
```

- 3. Visualization

- 3.1 Who Attended (Founders, Talents, Recruiters)?

```python
top10_attendees = df[df['describe_attendee'] != "No Input"]['describe_attendee'].value_counts().nlargest(10)

top10_attendees.sort_values().plot(kind='barh', title='Top 10 Attendees')
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.savefig('Images/Top10_Attendees', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

- 3.2  What do our top Attendees want?

```python
df['conference_expectation'].unique()
```

```python
df_filtered = df[df['conference_expectation'] != "No Input"].copy()

def categorize_expectation(text):
    if text == 'No Input':
        return None
    categories = {
        'Networking & Connections': ['network', 'connect', 'collaboration', 'relationships', 'meet', 'founder', 'investor', 'leader'],
        'Learning & Insights': ['learn', 'insight', 'trend', 'knowledge', 'info', 'learn from', 'insights into'],
        'Business Development & Funding': ['funding', 'raise funds', 'scale', 'business', 'investors', 'vc', 'sme', 'agrologistics', 'partner'],
        'Brand Visibility & Exposure': ['exposure', 'visibility', 'brand', 'attention'],
        'Personal & Professional Growth': ['skills', 'leadership', 'professional growth', 'expert', 'position myself'],
        'General Interest / No Specific Goal': ['find out more', 'no specific goal', 'general interest', 'get a view'],
        'Strategic / Ecosystem Analysis': ['consensus', 'market', 'ecosystem', 'prediction', 'market trends', 'analysis']
    }
    
    # Normalize the input text for easier matching
    normalized_text = text.lower().strip()
    
    # Iterate through categories and their keywords to find a match
    for category, keywords in categories.items():
        if any(keyword in normalized_text for keyword in keywords):
            return category
            
    # If no category is matched, return a default label
    return "Uncategorized"

df_filtered['attendee_category'] = df_filtered['conference_expectation'].apply(categorize_expectation)
```

```python
df_filtered = df_filtered.explode('attendee_category')

category_counts = df_filtered['attendee_category'].value_counts()
print(category_counts)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    y='attendee_category', 
    data=df_filtered, 
    order=category_counts.index, 
    palette='viridis'  # valid palette
)
plt.title("Attendee's Needs")
plt.xlabel('Count')
plt.ylabel('Attendee Category')
plt.savefig('Images/Attendees_Needs', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
